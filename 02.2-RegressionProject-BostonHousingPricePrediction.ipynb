## Dataset Information

Boston dataset was collected in 1978 with 506 entries and 14 attributes or features for home in the suburbs in Boston



<b><u>Attributes Information</b></u> 
<br>

1. <b>CRIM</b>:Per capita crime rate by town
2. <b>ZN</b>: Proportion of residential land zoned for lots over 25,000 sq.ft.
3. <b>INDUS</b>: Proportion of non-retail business acres per town
4. <b>CHAS</b>: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
5. <b>NOX</b>: Nitric oxides concentration (parts per 10 million)
6. <b>RM</b>: Average number of rooms per dwelling
7. <b>AGE</b>: Proportion of owner-occupied units built prior to 1940
8. <b>DIS</b>: Weighted distances to five Boston employment centres
9. <b>RAD</b>: Index of accessibility to radial highways
10. <b>TAX</b>: Full-value property-tax rate per \$10,000
11. <b>PTRATIO</b>: Pupil-teacher ratio by town
12. <b>B</b>: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
13. <b>LSTAT</b>: % lower status of the population
14. <b>MEDV</b>: Median value of owner-occupied homes in $1000's

## Exploratory Data Analysis

We will start by creating a box plots. Box plots are used to indentify the outliers. If there are many outliers, we need to ignore the outliers for better training of the model. We will create the box plot for all the attibutes using a loop

```python
# Creating the box plot

fig, ax = plt.subplots(ncols = 7, nrows = 2, figsize = (20, 10))
index = 0
ax = ax.flatten()
colors = ["#457b9d"]

for col, value in housingdata.items():
  sns.boxplot(y = col, data = housingdata, ax = ax[index], palette = colors, boxprops=dict(alpha=.5))
  index += 1

plt.tight_layout()
```

Here we have so many outliers in the columns: CRIM, ZN & B. Since there many  ourliers, they won't follow the uniform distribution.<br>
If we drop these rows, we would lose a lot of information.<br>
In order to make them a normal distribution, we would convert these values to log transformation or min-max normalization - to minimize the outliers.<br>

we will now look at the distribution of the datapoint

```python
# plot the dist plot

fig, ax = plt.subplots(ncols = 7, nrows = 2, figsize = (20, 10))
index = 0
ax = ax.flatten()

for col, value in housingdata.items():
  sns.distplot(value, ax = ax[index])
  index += 1

plt.tight_layout()
```

## Min-Max Normalization

```python
# Creating a list of columns which we will be normalising
cols = ["CRIM", "ZN", "TAX", "B"]
```

```python
# All the values in the column will be in the range of 0 - 1 when we do the min max normalisation. 
for col in cols:
  # find min and max for that column
  minnum = min(housingdata[col])
  maxnum = max(housingdata[col])
  housingdata[col] = (housingdata[col] - minnum)/(maxnum - minnum)
```

```python
# Displaying the distplot again

fig, ax = plt.subplots(ncols = 7, nrows = 2, figsize = (20, 10))
index = 0
ax = ax.flatten()

for col, value in housingdata.items():
  sns.distplot(value, ax = ax[index])
  index += 1

plt.tight_layout()
```

```python
# Standardrisation of the columns

# We will import the required library
from sklearn import preprocessing
scalar = preprocessing.StandardScaler()

# fit our data
scaled_cols = scalar.fit_transform(housingdata[cols])
scaled_cols = pd.DataFrame(scaled_cols, columns = cols)
scaled_cols.head()
```

```python
# adding this data back to our original dataframe
for col in cols:
  housingdata[col] = scaled_cols[col]

# Plotting the distribution again
fig, ax = plt.subplots(ncols = 7, nrows = 2, figsize = (20, 10))
index = 0
ax = ax.flatten()

for col, value in housingdata.items():
  sns.distplot(value, ax = ax[index])
  index += 1

plt.tight_layout()
```
