## Introduction to Regression Analysis 


Regression analysis examines the relationship between dependent (Outcome/ Response variable) variable Y and independent/ explanatory variable(s) X 
We summarise the relationship between X and Y using the line of best fit. 
The regression model estimates the slope of this line. The slope is often called as parameter or coefficient. 

Regression Analysis helps make predictions. because regression predicts the parameter that relates independent variable with the dependent variable.  
<br>It is useful when quantitaive evidence is needed to answer a particular question
it allows for testing a hypothesis and making predictions. 
Regression models helps estimate parameter and then they can be used to calculate new custom statistics. 

## Types of Regression Models

Simple Regression: When Y is related only 1 independent variable 
Multiple Regression: when Y is regressed against many independent variables (Most important). The multiple regression estimates the plane of best fit. 

There are different type of regression model classified depending upon the nature of the data (Cross Sectional/ Repeated) and the type of dependent variable (Continuous/ categorical)


For cross sectional Data:

when dependent variable is
1. Continuous  - we use OLS (linear)
2. Binary - Logit/Probit (non-linear)
3. Ordered categories - Ordered Logit (non-linear)
4. Nominal categories - multinomial Logit (non-linear)
5. Count - Poisson Binomial (non-linear)

For over time Repeated Data:

when the dependent variable is
1. Continuous: Linear Panel Model & Non-Linear Panel Model 


Note:<br>
1. Cross-sectional Data: is the data collected at a single point in time. We are not interested in the change of data over time.

## Concept of Linear Regression Model

Linear Regression is the most popular regression method. It assumes that the variable parameter in a linear way. For applying the linear regression, the dependent variable must be a continuous variable/ or close to continuous variable. 
The main ideas behind the linear regression is that we use the least squares to fit a line through the data (also known as OLS - Ordinary Least Squares). <b>OLS</b> is the most popular regression Technique.<br>

Relationship between the variables:


$$
  Y = m_1X_1 + m_2X_2 + m_3X_3 + C
$$


<br>where:<br>
><b>Y</b>: is the Target variable which we are trying to predict. <br>
><b>X:</b> are the independent variables which might have a relationship with the dependent variable.  
><b>m:</b> variable paramter/ Coefficient we are estimate through the model. 

<b>Residuals</b>: the distance between the line and the data point. 

<b> SS(mean)</b>: Sum of squares around mean. (data - mean)<sup>2</sup> 
variation around the mean = ss(mean)/n where n = sample size. 
<b>ss(fit></b> sum of squares around the least square line 
(data - line)<sup>2</sup>
varianc can be defined as the average of sum of squares. 


Regression is all about lines and how you fit it through the data. 

R-squred: it is a value in between 0 - 1. A higher value indicates a better fitting model. 
No of observation: highlights the sample size of the regression model. It is very important data point, as there are missing data whoch might reduce the observation count 
F-Statistics: determines whether the model has any predictive power. Higher value means that the model has more predictive power. 
Root MSE: is a way of measuring the total size of the residual. 
Coefficient: which determines the estimated relationship each variable has with the target variable. 
Standard erorr: a statistics determining the degree of accracy the coefficients are estimated ti. Lower values indicate higher accuracy, with the standard error, you can determine to what extend the estimated cofficient are statistically significant. 
t-statistics: for observation count above 100, t-stat over 1.96 are generally significant. 
p-value below 0.05 are statistically significant. 

Regression analysis is only correlation not the causation. 
causation is always interpretated by the end user. Causality requires: Theoretical reasoning, common sense and logic.

## Interpretation of Logs in Regression Analysis

#### Level - Level 

Dependent variable is Y, Independent Variable is X <br>
$$
ΔY = β_1ΔX
$$

A unit change in X causes a β<sub>1</sub> unit change in Y

<br>

#### Level - Log 

Dependent variable is Y, Independent Variable is  ln(X) <br>
$$
 \% ΔY = (\frac{β_1}{100})\%  ΔX
$$

A 1\% change in X causes a β<sub>1</sub>/100  unit change in Y

<br>

#### Log - Level 


Dependent variable is ln(Y), Independent Variable is  X <br>
$$
 \% ΔY = ({β_1}*{100})\%  ΔX
$$

A 1\% change in X causes a β<sub>1</sub> * 100  unit change in Y

<br>
